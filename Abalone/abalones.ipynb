```python
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:/git/ml/Abalone/data/playground-series-s4e4/train.csv')
df
```

```python
df.describe()
```

```python
df['Sex'] = df.Sex.map( {'F':0 , 'I':1, 'M':2} )
df
```

```python
X_train = df[['Sex','Length','Diameter','Height','Whole weight','Whole weight.1','Whole weight.2','Shell weight']]
y_train=df['Rings']
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
df_test = pd.read_csv('E:/git/ml/Abalone/data/playground-series-s4e4/test.csv')
df_test['Sex'] = df_test.Sex.map( {'F':0 , 'I':1, 'M':2} )
df_test
```

```python
X_test = df_test[['Sex','Length','Diameter','Height','Whole weight','Whole weight.1','Whole weight.2','Shell weight']]
y_test=df_test['Rings']
y_pred = regr.predict(X_test)
```

```python
y_pred
```

```python
print("Coefficients: \n", regr.coef_)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```
